## Read from bronze

```python
df = spark.table("workspace.bronze.erp_loc_a101")
df.display()
```

### Standarization and trim the data

```python
query = """
    SELECT
        REPLACE(CID,'-','') AS new_cid,
        CASE 
            WHEN UPPER(TRIM(CNTRY)) IN ('US', 'USA') THEN 'United States'
            WHEN UPPER(TRIM(CNTRY)) = 'DE' THEN 'Denmark'
            WHEN CNTRY IS NULL OR TRIM(CNTRY) = '' THEN 'n/a'
            ELSE TRIM(CNTRY)
        END CNTRY
    FROM workspace.bronze.erp_loc_a101;
"""

df = spark.sql(query)
df.display()
```

### Load into sliver layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.erp_loc_a101")
```
