## The Transformation Logic

```python
query = """
    SELECT
        pi.product_id,
        pi.category_id,
        pi.product_key,
        pi.product_name,
        pi.product_cost,
        pi.product_line,
        pc.CAT AS cateogry,
        pc.SUBCAT AS sub_category,
        pc.MAINTENANCE AS Maintenance,
        pi.product_start_date,
        pi.product_end_date
    FROM silver.crm_prd_info pi
    LEFT JOIN silver.erp_px_cat_g1v2 pc
    ON pi.category_id = pc.ID
"""

df = spark.sql(query)
df.display()
```

### Write into gold layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
