## Transformation Logic

```python
query = """
    SELECT
        sd.sls_order_number,
        dp.product_key,
        dc.customer_id,
        sd.sales,
        sd.quantity,
        sd.price,
        sd.sls_order_date,
        sd.sls_shipping_date,
        sd.sls_due_date
        FROM silver.crm_sales_details sd
    LEFT JOIN gold.dim_products dp
    ON sd.sls_product_key = dp.product_key
    LEFT JOIN gold.dim_customers dc
    ON sd.sls_customer_id = dc.customer_id
"""

df = spark.sql(query)
df.display()
```

### Write into gold layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
