## The Transformation Logic

```python
query = """
    SELECT
        ci.customer_id,
        ci.customer_key,
        ci.firstname,
        ci.lastname,
        ci.marital_status,
        CASE 
            WHEN ci.gender != 'n/a' THEN TRIM(ci.gender)
            ELSE COALESCE(ci.gender, 'n/a')
        END AS gender,
        cz.BDATE AS bdate,
        la.CNTRY AS country,
        ci.created_at
    FROM workspace.silver.crm_cust_info ci
    LEFT JOIN silver.erp_cust_az12 cz
    ON ci.customer_Key = cz.CID
    LEFT JOIN silver.erp_loc_a101 la
    ON ci.customer_key = la.new_cid
"""

df = spark.sql(query)
df.display()
```

### Write into gold layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_customers")
```
