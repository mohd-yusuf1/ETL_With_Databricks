### Reading CSV File And Ingest Data Into Bronze Layer Using Python

```python
INGESTION_CONFIG = [
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_system/source_crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_system/source_crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_system/source_crm/sales_details.csv",
        "table": "crm_sales_details"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_system/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_system/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_system/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze.{item['table']}")

    df = (
        spark.read.option("header","true")
        .option("inferschema", "true")
        .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze.{item['table']}")
    )
```

### Droping All Tables In One Go Using Python

```python
for i in INGESTION_CONFIG:
    print(f"Droppin Table {i['table']}")
    db_query = f"DROP TABLE workspace.bronze.{i['table']}"

    spark.sql(db_query)
```

### Fetching Data From Each Table In One Go Using Python

```python
for each in INGESTION_CONFIG:
    print(f"Fetching From Table {each['table']}")
    db_query2 = f"SELECT * FROM workspace.bronze.{each['table']} LIMIT 3"

    df = spark.sql(db_query2)
    df.display()
```
